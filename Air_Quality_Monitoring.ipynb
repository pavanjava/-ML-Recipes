```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
%matplotlib inline
```

```python
df = pd.read_csv("C:\\Pavans\\#{ML-DL}\\datasets\\india-air-quality-data\\data.csv",encoding='latin1')
```

```python
df.head()
```

```python
df = df.fillna(0)
df.head()
```

```python
df.describe()
```

```python
so2_visualization = df.groupby(['state']).mean()['so2'].sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(14,14))
sns.barplot(x = 'so2', y = so2_visualization.index,data = so2_visualization,palette='inferno')
plt.title('So2 concentration at the highest level statewise')
plt.xlabel('SO2 level',fontsize = 15)
plt.ylabel('State',fontsize = 15,rotation = 'horizontal')
```

```python
no2_visualization = df.groupby(['state']).mean()['no2'].sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(14,14))
sns.barplot(x = 'no2', y = no2_visualization.index,data = no2_visualization,palette='inferno')
plt.title('No2 concentration at the highest level statewise')
plt.xlabel('NO2 level',fontsize = 15)
plt.ylabel('State',fontsize = 15,rotation = 'horizontal')
```

```python
hyd_df = df[df.location == 'Hyderabad'].fillna(0)
```

```python
hyd_no2_levels = hyd_df.groupby(['location_monitoring_station']).mean()['no2'].sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(14,14))
sns.barplot(x = 'no2', y = hyd_no2_levels.index,data = hyd_no2_levels,palette='inferno')
plt.title('No2 concentration at the highest level statewise')
plt.xlabel('NO2 level',fontsize = 15)
plt.ylabel('State',fontsize = 15,rotation = 'horizontal')
```

```python
hyd_so2_levels = hyd_df.groupby(['location_monitoring_station']).mean()['so2'].sort_values(ascending=False).to_frame()
```

```python
plt.figure(figsize=(14,14))
sns.barplot(x = 'so2', y = hyd_so2_levels.index,data = hyd_so2_levels,palette='inferno')
plt.title('SO2 concentration at the highest level statewise')
plt.xlabel('SO2 level',fontsize = 15)
plt.ylabel('State',fontsize = 15,rotation = 'horizontal')
```

```python
fig = plt.figure(figsize = (12,3))

plt.subplot(1,4,1)
sns.distplot(df['so2'],color='r')

plt.subplot(1,4,2)
sns.distplot(df['no2'],color='b')

plt.subplot(1,4,3)
sns.distplot(df['spm'],color='g')

plt.subplot(1,4,4)
sns.distplot(df['rspm'],color='y')
```

```python
#Cleaning the column values 'type'
df['type']=df['type'].replace(['Sensitive Area','Sensitive'],(['Sensitive Areas','Sensitive Areas']))
df['type']=df['type'].replace(['Industrial Area','Industrial'],(['Industrial Areas','Industrial Areas']))
df['type']=df['type'].replace(['Residential','Residential and others'],(['Residential, Rural and other Areas','Residential, Rural and other Areas']))
```

```python
plt.figure(figsize = (8,6))
sns.countplot(df['type'])
plt.xticks(rotation=90)
```

```python
fig,ax = plt.subplots(1,2,figsize = (12,2))
ax1,ax2 = ax.flatten()
cnt = df.groupby(['type']).max()['no2'].sort_values(ascending = False).to_frame()[:50]
sns.barplot(x = cnt['no2'],y = cnt.index,ax = ax1 ,palette= 'winter')
ax1.set_title('No2 levels Comparison with Area Types')

cnt = df.groupby(['type']).max()['so2'].sort_values(ascending = False).to_frame()[:50]
sns.barplot(x = cnt['so2'],y = cnt.index,ax = ax2 ,palette=  'inferno')
ax2.set_title('So2 levels Comparison with Area Types')
plt.subplots_adjust(wspace = 0.8)
```
